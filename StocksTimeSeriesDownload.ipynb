```python
import pandas as pd

df = pd.read_csv('/content/StockMarketML/sp500_companies.csv')
```

```python
df.Symbol
```

```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
import yfinance as yf

%matplotlib inline
```

```python
# Now let's define a range, that means, from when until when we want the data:
start_date= dt.datetime(1995, 1, 1) # the format is YYYY-MM-DD
end_date= dt.datetime(2024, 6, 1)
```

```python
stocks_timeseries = pd.DataFrame()
for x in df.Symbol:
  df1 = yf.Ticker(x.strip()).history(start = start_date,end = end_date)
  stocks_timeseries = pd.concat([stocks_timeseries, df1.rename(columns = {'Open':x})[x]], axis = 1)
```

```python
stocks_timeseries.tail(5)
```

```python
stocks_timeseries.to_csv('stocksTimeSeries.csv')
```
