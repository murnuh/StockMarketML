```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
import yfinance as yf

%matplotlib inline
```

```python
# Now let's define a range, that means, from when until when we want the data:
start_date= dt.datetime(1995, 1, 1) # the format is YYYY-MM-DD
end_date= dt.datetime(2024, 6, 1)
```

```python
df2 = yf.Ticker('CL=F').history(start = start_date,end = end_date)
```

```python
df2.tail()
```

```python
MacroStocks_timeseries = pd.DataFrame()
ticker = ['CL=F','^SPX','DX-Y.NYB','TLT']
for x in ticker:
  df1 = yf.Ticker(x).history(start = start_date,end = end_date)
  MacroStocks_timeseries = pd.concat([MacroStocks_timeseries, df1.rename(columns = {'Open':x})[x]], axis = 1)
```

```python
MacroStocks_timeseries.to_csv('MacroStocksTimeSeries.csv')
```
